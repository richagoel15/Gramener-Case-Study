### Derived Data Calculation and Analysis

```python
# Check unique value for int_rate column
df.int_rate.unique()[1:10]
```

```python
## Convert int_rate and revol_util column to numeric for further analysis 

df.int_rate = pd.to_numeric(df.int_rate.str[:-1])
```

```python
##Devide interest rate in three broad category , Low , Medium and high-
def rate_range(rate):
      
    if (rate in range(1,9)): 
        int_range = "Low"
    elif(rate > 9) and (rate <= 15):  
        int_range = "Medium"
    
    else:  
        int_range = "High"
        
    return  int_range
```

```python
 #Calculating interest rate  Range column by calling rate_range function
df['rate_range'] = df['int_rate'].apply(rate_range)
```

```python
#Check unique value for revol_util column
df.revol_util.unique()[1:10]
```

```python
# Convert revol_util to numeic value for further analysis 
df.revol_util = pd.to_numeric(df.revol_util.str[:-1])
```

```python
## check DTI column
df.dti.unique()
```

```python
df
```

```python
# Check unique value for employer length
df.emp_length.unique()
```

```python

# Convert employee length to numeric for further analysis :-

df.emp_length = df.emp_length.str.replace(' years','')
df.emp_length =df.emp_length.str.replace('+','')
df.loc[df.emp_length == '< 1 year','emp_length'] = 0
df.loc[df.emp_length == '1 year','emp_length'] = 1

#df.emp_length.fillna(-1, inplace=True)
df.emp_length = pd.to_numeric(df.emp_length)
```

```python
#Convert all Date fields ( issue_d , last_pymnt_d and last_credit_pull_d ) to same date format

df.issue_d = pd.to_datetime(df.issue_d, format='%b-%y')

df.last_pymnt_d = pd.to_datetime(df.last_pymnt_d, format='%b-%y')
df.last_credit_pull_d = pd.to_datetime(df.last_credit_pull_d, format='%b-%y')
```

```python
## Add a new column and derive earlist Credit year from  earliest_cr_line
def convert_date(row):
    trim_date1 = row[-2:]
    if (trim_date1 in ['00','01','02','03','04','05','06','07','08','09','10','11','12','13','14','15','16','17','18']):
           credit_year= '20' +  trim_date1
    else:  
           credit_year= '19' +  trim_date1
            
    return credit_year
```

```python
df['Credit_ag_year'] = df['earliest_cr_line'].apply(convert_date)
df['Credit_ag_year'] = pd.to_numeric(df.Credit_ag_year)
```

```python
##Add a new column Loan_Amnt_Range 

labels = ["{0} - {1}".format(i, i + 2800 ) for i in range(0, 28000, 2800)]

df['loan_Amnt_range'] = pd.cut(df.loan_amnt, range(0, 29000, 2800), right=False, labels=labels)
```

```python
## Add a column for Annual income range

labels = ["{0} - {1}".format(i, i + 12000) for i in range(0, 120000, 12000)]
labels

df['Annual_inc_Range'] = pd.cut(df.annual_inc, range(0, 130000, 12000), right=False, labels=labels)
```

```python
## Add a new column for no of inquiry in last 6 month 
labels = ['0-2','2-4','4-6','6-8']

labels
df['inq_range'] = pd.cut(df.inq_last_6mths, range(0, 10, 2), right=False, labels=labels)
```

```python

## Add a new column Revol_uti_range for revol_util
labels = ["{0} - {1}".format(i, i + 10 ) for i in range(0, 100, 10)]


labels
df['revol_util_range'] = pd.cut(df.revol_util, range(0, 105, 10), right=False, labels=labels)
print(df)
```
